# 1. Motivation

__Narrow__: If we want to find a statistical twin or create a statistically randomized controlled trial, we need to measure the appropriate distance. 

For example, a statistical twin as similar to the unit of interest as possible in terms of the covariates X that matter for the outcome y. It will usually not be the same as X, so how do we measure how different it is? The answer lies in an appropriate metric, distance, divergence or contrast function.

The same is true for "pseudo" randomization. Here we want to make sure that the treatment and control are equally distributed amongst the relevant Xs, so we need to have a metric to measure how different these distributions are.


__Wider__: Any type of machine learning involves a distance function. If we try to find the best fitting line, how do we measure the distance form this line and thus how do we fit the line? If we measure how similar images generated by GANs are, how do we measure the difference between distributions and so on..



__Note that each distance metric or contrast function makes an assumption about the underlying structure of the problem__ 


__There is no general right or wrong, but only a fit with the underlying problem!!!!__

# 2. Definition (statistics)


__Remember the definition of a metric, also known as distance or contrast function__

Metrics [ edit ]
A metric on a set $X$ is a function (called the distance function or simply distance)
$d: X \times X \rightarrow R^{+}$ (where $R^{+}$ is the set of non-negative real numbers). For all $x, y, z$ in $X$, this function is required to satisfy the following conditions:
1. $d(x, y) \geq 0 \quad$ (non-negativity)
2. $d(x, y)=0$ if and only if $x=y$ (identity of indiscernibles. Note that condition 1 and 2 together produce positive definiteness)
3. $d(x, y)=d(y, x) \quad$ (symmetry)
4. $d(x, z) \leq d(x, y)+d(y, z) \quad$ (subadditivity / triangle inequality).

In statistics and information geometry, divergence or a contrast function is a function which establishes the "distance" of one probability distribution to the other on a statistical manifold. The __divergence is a weaker notion than that of the distance__, in particular the divergence need not be symmetric (that is, in general the divergence from p to q is not equal to the divergence from q to p), and need not satisfy the triangle inequality.



Suppose $S$ is a space of all probability distributions with common support. Then a divergence on $S$ is a function $D(\cdot \| \cdot): S \times S \rightarrow R$ satisfying
1. $D(p \| q) \geq 0$ for all $p, q \in S$
2. $D(p \| q)=0$ if and only if $p=q$
