<img src="https://gym.openai.com/assets/docs/aeloop-138c89d44114492fd02822303e6b4b07213010bb14ca5856d2d49d6b62d88e53.svg" width=45%>

```python
# sample code from the documentation
# !pip install gym
import gym
env = gym.make('CartPole-v0')
for i_episode in range(20):
    observation = env.reset()
    for t in range(100):
        env.render()  # this can not work on Colab
        print(observation)
        # choose the next action
        #env.step(env.action_space.sample()) # take a random action
        # this should be done by the RL agent
        action = env.action_space.sample()
        # don one step in the environment
        observation, reward, done, info = env.step(action)
        if done:
            print("Episode finished after {} timesteps".format(t+1))
            break
env.close()
```

* unified interface over all the environments
* rl algorithm implementations tend to be compatible with it
* contains a vast amount of environments: Atari, classic, robot control
