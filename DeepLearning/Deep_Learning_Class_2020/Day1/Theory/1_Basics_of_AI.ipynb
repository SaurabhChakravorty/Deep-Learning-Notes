# Learning vs modeling

## What is a model?

A model is an artificial conceptual and/or physical structure that:
* captures / stores essential property or working mechanism of real phenomena (Thus has to be isomorphic, homomorphic or at least similar to the original)
* makes phenomena reproducible / predictable / studyable
* focuses attention / discards unnecessary details / noise

More about the semantics of models [here](https://plato.stanford.edu/entries/models-science/#RepModIModPhe)

### Models can be parametric vs non-parametric

**All models that have parameters that can be adjusted according to observations/data are called parametric. Models without such parameters are non-parametric.**

**When we speak about "models" a in machine learning sense, we talk about structures with modifiable parameters that (can) encode knowledge ("learn") through this modification process ("learning", "storage" or "optimization") and retrieve/apply those in relevant new situations, on new "input" ("retrieval", "inference" or "prediction").**

This justifies why the school emphasizing learning and the one emphasizing knowledge could arise.

This knowledge can be represented in terms of discrete symbols and their combinations / relations (like a knowledge base) or as "distributed representations" (like a set of coefficients for a regression model). This can affect the human "interpretability" of the model itself.

## [Empirical risk minimization](https://en.wikipedia.org/wiki/Empirical_risk_minimization)

We are talking about "empirical risk minimization", that is to say we have a dataset which we consider being drawn from an underlying data distribution, from which we would like to learn something. We do not know the true, underlying distribution (the true, underlying relationships), so we estimate the model that best represents the relationships presented in the data. In order to do so, we minimize the "empirical risk", that is we select the model that most likely seems to be true given the empirical data we have obtained.

In a sense machine learning is all about generalization, finding relationships from data that are generalizable and thus help us to make accurate predictions. When we talk about empirical risk minimization we consider how we can find the best model of underlying empircical data that we are considering. There will be other factors that determine whether a model generalizes well, that do not just directly apply to our data-set and we will come back to those throughout the course.

Empirical risk minimization, i.e. selecting a model that best describes relationships in a particular data set, requires us to define what we take as performance of the model (i.e. "best"). Any empirical risk minimization is in respect to the performance metric we have chosen for a model, and it is important to keep this in mind.

The risk is then defined as the expectation of the loss function, i.e. the error that the estimated function produces in terms of the relationships between variables. We are henceforth trying to minimize this error.

The expectation of the loss function cannot usually be calculated, as we do not know the true underlying relationship. however, we can calculate an approximation, called empirical risk.

The empirical risk minimization principle states that the learning algorithm should chose the model (hypothesis) that minimizes the empirical risk.

Thus following the empirical risk minimization principle requires solving and optimization problem.

We typically measure the performance of our models in some kind of **error metric**, which reflects direct performance on the task (we will talk about this later in detail), but it is important to understand, that regularly we will set more complex **"objectives"** for the model, that not just contain **some measure of cost, but also other boundary conditions** (complexity, conformance to some constraints, etc.), and during our **optimization process** we wish to find the best model for this objective.

At the end of the day the model only optimizes what objectives we have set for it and it is essential to understand the different assumptions we make when the learning process is defined. In a way crafting those assumptions to fit with the underlying problem is the key to machine learning.

## Bayesian and frequentist statistics in a nutshell

Since we will be referring to statistical concepts during the course, we have to consider the two "schools" of statistics we may encounter: the Bayesian and frequentist approaches to statistics.

### Probability
The two approaches are based on two fundamentally different interpretations of probability, although, of course, both accept the same mathematical theory based on Kolmogorov's axioms.

#### Frequentist probability 

For frequentists, the probability of an event is its long-term frequency in a large number of trials. Only events that  are the outcomes of well defined, repeatable random processes ("random experiments") have probabilities.
#### Bayesian probability

Bayesians, in contrast, interpret probabilities as "degrees of beliefs", and these beliefs can be not only about repeatable random events but also general hypotheses, single unrepeatable events (e.g. that Trump will be reelected in 2020) etc.

As degrees of beliefs, Bayesian probabilities can change in time, so the laws of probability on this interpretation are concerned with the constraints on maintaining coherent and consistent degrees of beliefs for a subject. One of the central questions is that of properly _updating_ degrees of beliefs/probabilities in the face of new evidence. For Bayesians,  Bayes's theorem/rule:
$$ 
P(A \vert B) = \frac{P(B \vert A)\cdot P(A)}{P(B)}
$$
is the central result for this topic: When applied to a hypothesis $H$ and evidence $E$, it becomes 
$$
P(H \vert E) = \frac{P(E \vert H)\cdot P(H)}{P(E)}
$$
which, interpreted as an update rule says that if your **prior** degree of belief in a hypothesis and an evidence are $P(H)$ and $P(E)$, and your assessment of the probability (**likelihood**) of $E$ on the condition that $H$ is $P(E\vert H)$ then, when you learn that $E$ is in fact the case, your updated **posterior** degree of belief regarding $H$ should be $\frac{P(E \vert H)\cdot P(H)}{P(E)}$.

### Parameter estimation
The two approaches to probability have important implications for parameter estimation as well. 

#### Bayesian
If we have a sample from a distribution and want to estimate the parameters of the distribution, then Bayesians can simply treat the distribution parameters as variables with a certain **prior** probability **distribution** (a distribution over the available hypotheses about the parameter's value) and calculate the **posterior distribution** for the parameter's value on the basis of the sample as evidence using Bayes' rule. This posterior distribution, in turn, can be used to provide a point estimate (e.g. choosing the value with the maximal posterior probability) or an interval estimate (the value is in a certain range with a certain probability) for the parameter.

#### Frequentist
Frequentists, in contrast, cannot treat the parameter values as variables with a probability distribution, since they are single concrete values, not the outcomes of repeatable random experiments. Accordingly, they treat the **sample** as random (for the Bayesian the sample is fixed!) and try to find estimators (functions from samples to estimated parameter values) whose distribution over the samples is useful for point and interval estimates. Important positive estimator properties are

+ being **unbiased**: the expected value of the estimator coincides with the true parameter value
+ being **consistent**: as the sample size increases, the sampling distribution of the estimator becomes increasingly concentrated at the true parameter value.

One of the frequently used estimators is the Maximum Likelihood Estimate (MLE) which coincides with the Bayesian Maximum A Posteriori (MAP) point estimate for the sample assuming a uniform prior distribution.

### Variance in sampling

- ["Martingale betting"](https://en.wikipedia.org/wiki/Martingale_(betting_system) )
    - "Always double the bet if you have lost"
    - "...a gambler with infinite wealth will, almost surely, eventually flip heads, the martingale betting strategy was seen as a sure thing by those who advocated it."
    
    
- 100 random coinflips, we would expect to have a fairly low amount of consequtive heads or tails (1-s or 0-s), since it's a "fair" coin, and in the margin it should come up at 50-50%.
    - The problem is only just it does not in practice! :-)
    -  "... combined with the fact that strings of consecutive losses actually occur more often than common intuition suggests, can bankrupt a gambler quickly."

```python
import numpy as np
import matplotlib.pyplot as plt

colors = ["r","g","b","y"]

seeds = [1,2,3,4]

for counter in range(len(seeds)):
    state = np.random.RandomState(seeds[counter])

    random_array = state.choice([0,1],(100))

    c = np.concatenate([np.cumsum(c) if c[0] == 1 else c for c in np.split(random_array, 1 + np.where(np.diff(random_array))[0])])
    counts = {}

    previous_zero = True

    for i in c[::-1]:
        if i !=0:
            if previous_zero == True:
                try:
                    counts[i]+=1
                except:
                    counts[i]=1
            previous_zero = False
        else:
            previous_zero = True

    x = list(counts.keys())
    y = list(counts.values())

    plt.scatter(x,y,c=colors[counter])
    
    #Watch the list comprehension at work! http://www.pythonforbeginners.com/basics/list-comprehensions-in-python
    plt.legend(["Draw "+str(i) for i in seeds])

print()
print("Length of draws: ",len(random_array))
    
plt.title("Number of consecutive zeros from a random draw")
plt.show()

plt.title("Proportion of ones and zeros")
plt.hist(random_array)
plt.show()

print("2 raised to the 7 is 128 times the original bet amount - and I may get even more unlucky...")
```

### Insufficient data

If I only draw from one specific part of the data distribution and set up a hypothesis based on that, I can become unable to say anything relevant about the "other" parts, which could be quite radically different!


```python
import matplotlib.pyplot as plt
import pylab

pylab.rcParams['figure.figsize'] = (12, 4)

def generate_poly_base(degree, min=-0.8, max=1.1, num=500, lastneg=True, regularize=False):

    coefs = np.ones(degree+1)
    if regularize:
        coefs = [co*0.000000001 for co in coefs]
        
        coefs[0] = 1.0
        coefs[1] = 1.0
        coefs[2] = 1.0

    if lastneg:
        coefs[-1] = -1.0*coefs[-1]
    x1=np.linspace(min,max,num)
    x2=np.zeros(x1.shape)
    
    for d in range(degree+1):
        x2+=coefs[d]*(x1**d)
    
    x2=x2+x2*np.random.uniform(low=-0.2, high=0.2,size=len(x2))
    
    return np.array([x1,x2])


print("What is the difference between the two datasets? - Well, on this part, not much...")

fig, ((ax1, ax2)) = plt.subplots(nrows=1, ncols=2)

m1 = generate_poly_base(5)
ax1.plot(m1[0],m1[1])


m2 = generate_poly_base(2, lastneg=False)
ax2.plot(m2[0],m2[1])

plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.75, hspace=None)
plt.show()
plt.close()


pylab.rcParams['figure.figsize'] = (14, 4)
print("Albeit if we look at it on a further part!!")
fig, ((ax3, ax4)) = plt.subplots(nrows=1, ncols=2)

m1 = generate_poly_base(5,max=2.0)
ax3.plot(m1[0],m1[1])
ax3.axvline(x=1.1, c="r")
ax3.set_title("Fifth degree polynomial")


m2 = generate_poly_base(2,max=2.0 ,lastneg=False)
ax4.plot(m2[0],m2[1])
ax4.axvline(x=1.1, c="r")
ax4.set_title("Second degree polynomial")

plt.show()


```

### Problems with repeated hypothesis testing

The traditional "goodness of fit" type statistics based on *p* values only tell us how unlikely it is that a given hypothesis is true by chance. We saw, that if we repeatedly test hypothese after each-other, "significant p-values" could and will occur by chance or due to local relationships in the data. These relationships are not useful for prediction!


Reproducibility of scientific studies:
- Clinical sciences: : > 50% of studies in the clinical sciences may not be reproducible (Ioannidis, 2005)
- Top psychological journals: Open Science Project: 39 out of 100 studies reproduce (reproduced 100 studies)
- Strategic management: 24-40% of studies not be reproducible (Goldfarb & King 2016)

An example:
- N hypotheses, all independent
- Assume each of these hypotheses is tested on a separate  subset of data 
- Probability of rejecting at least one true null hypothesis is p = $1 - (1 - α)^n$
- n = 1 and α = 0.01 we have, p = 0.01. 
- n=100, α = 0.05     => p = 0.9941 
- Some “proven results” are wrong”

This is also quite unpleasant in general for matters of scientific progress and truth.

More:
- [Testing too many hypotheses](http://www.spencergreenberg.com/2011/10/testing-too-many-hypotheses/)
- [Why Most Published Research Findings Are False](http://journals.plos.org/plosmedicine/article?id=10.1371/journal.pmed.0020124)


## Conclusion

Even after considering the above caveats, we can conclude:

**Learning in machine learning context is using optimalization techniques to find optimal parameters (and sometimes structure) of models to maximize "explanatory" or "predictive" value with regard a data sampled from an unknown data distribution. The key question is how we can generalize from such data set, so that the relationships identified in this dataset are useful for making preditions in as many situations and across as many contexts and time frames as possible. This problem of "generalization" can be seen as one of the key drivers behind the success of machine learning. Some of its aspects are reasonably well understood, others are not understood well at all.**

# Paradigms of learning 

There are three main "paradigms" of leraning that are frequently employed in machine learning.

<img src="https://nowenlightenme.files.wordpress.com/2018/03/main-qimg-dc94548dc21042e3595af28d897d8a76-c.jpg?w=602&h=280&crop=1" width=600 heigth=600>

## Supervised learning

- Data is given as $(X, y)$ variable pairs, that is **input ($X$)** measurements and (target) **labels ($y$)**
- This is also called task driven learning, since the $y$ target value gives us a specific task, we have to predict $y$ from $X$ as well as possible
- This is a "direct feedback" setting, for all predictions we get feedback

Since this is the most widespread business setting, we will spend the majority of time discussing this, though it is worth noting that this setting is **highly dependent on large amounts of good quality labeled data**, which can make it's application difficult.

## Unsupervised learning

- Only data in form of $X$ mesurements is given, no labeling
- This is a "no feedback" scenario, we have no additional "teaching signal"
- We are aiming at finding hidden or latent structure in the data and represent / visualize it

This is a very desirable learning paradigm since it does not require hand labeling of data, but we are still at the early stages of exploring it's potential.

## Reinforcement learning

- We think in interactions of agents and environments (frequently in a simulation)
- The agent sometimes, quite infrequently receives some feedback signal from the environment after it carried out a "trial and error" style set of actions
- Learning of optimal behavior in a complicated environment is the main goal here

A somewhat distinct paradigm even inside AI, "if it works, it works great, if it fails, it fails big".

# Task types

The following broad categories are present in ML as problem or task formulations:

1. Visualization / representation

We would like to get an overview / insight into our data.

2. Clustering (similarity, retrieval)

Closely related to the previouls one, we would like to see discernible patterns inside the data and be able to measure similarities, thus enableing recommendation and retrieval.

3. Classification

The task is to assign pre-set categorical class labels to the data.

4. Regression / forecasting

The task is to produce a continuous numeric output estimate of a given numeric target measurement, quite often into the future (forecasting).

5. Control / optimization

Finding of optimum settings / actions of complex systems.

6. Generation

Creation of novel data instances based on prior leanring samples (natual language generation, image generation...)

**Note: careful consideration should be given to the formulation of the task, since often it is true that (nearly) the same business value can be realized, and the reformulation of the task setting decides the feasibility of the solution.**
