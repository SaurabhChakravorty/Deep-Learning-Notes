## Scope

There can be considerable variation in the motivation for building up chatbot systems. 

The general dichotomies are:

[Source](https://dzone.com/articles/chatbots-categories-and-their-limitations-1)

## Social vs. Task oriented

While the social bots serve the purpose of entertainment and social connection with the users, the more wide spread task oriented bots are having a specific (business) function, and their goal is solely to carry that out. 

## Open domain vs. closed domain

Bots can try to handle all types of dialog situations from different domains, but the successful ones often focus on a specific, narrow domain to have deep enough knowledge.

Often these types of bots are not compatible. 

"There is no such thing as a free lunch." [theorem](https://en.wikipedia.org/wiki/No_free_lunch_theorem)

Generally we will concern ourselves with **task oriented, closed domain bots**, since they have more business value, as well as the results in this field are way more pronounced.

### Analogies to "webforms"

<a href="https://img00.deviantart.net/9eb3/i/2015/122/6/f/web_form_elements_vol_2_by_sniperyu-d2f8y4w.jpg"><img src="https://drive.google.com/uc?export=view&id=1P5LcaeaCT19YV80dZsbehK-aFTaxs2ld" width=55%></a>

### Main structural elements:

- **"Dialog flow"**: The branches of conversations that are explicitly supported, and some "failover" logic to handle exceptions
- **"Intents"**: What action or service the user's wants? (A general utterance or a context points to "nodes" in the dialog graph (analogy: choose a specific webform from the menu)
- **"Entities"**: What parameters are to be extracted from the utterance? (Broader sense then in NLP, basicly the "form fields" with data types that are required for a service call)
- **"Responses"**: Format the results we got from the services and present it to the user in a human readable way. ("Generate answer".)



## Approaches to solution

### Rule based

The original concept of the "Chinese" room - funnily enough as a model for machine translation.

<a href="https://devanshmittal.files.wordpress.com/2012/07/images.jpg"><img src="https://drive.google.com/uc?export=view&id=1DtxopBJgn8jF3WXjMNsAey4BViAqlxR6" width=45%></a>

#### Solution structure:
- Pre-coded dialogue flow
- Keyword matching for intent detection (list of keywords + RegExp)
- Rule based entity extraction
    - RegExp again
    - Morphology based clues?
    - MAYBE POS or depdndency?

- Ontologies and rule engines can come in easily to help the production of answers
- Pre-configured "panels" for answers

#### Coverage problem

It is not difficult to see, that this approach suffers from a _huge_ coverage problem: Way too many variants to cover.
Illustration: Hungarian as language with rich morphology:

<a href="https://slideplayer.com/slide/8040825/25/images/3/Hungarian+morphology+Agglutinative+%28and+sometimes+inflectional%29.jpg"><img src="https://drive.google.com/uc?export=view&id=1mU_bYwu4l6DYzqnwJW7O2lN8ltXr9EnV" width=55%></a>

**Pure rule based systems are generally unscalable.**



### Flow + statistical models

It is easy to see how the advancement of NLP methods could help the chatbot workflow, especially with the advent of high quality language models (and higher structure models, like sentence embeddings). 

#### Solution structure:
- **Still** Pre-coded dialogue flow 
    - See the easy UI for flow buildup above
- Embedding based intent detection 
    - an utterance is just a query for a small set of intent "documents" after all...
- Embedding based and/or custom trained entity extraction model 
    - Typically not a complete NER parser, just based on some distance metric from known "entities", which can be of any data type
    - Naturally lexicalization and morphology, as well as POS and dependency can help is needed
- Pre-configured "panels" for answers, response comes from webservices (or in rare case knowledge bases

Let's remember: **distances and directions are systematic in an embedding space!!**

<a href="https://image.slidesharecdn.com/vectorland-151126121952-lva1-app6892/95/vectorland-brief-notes-from-using-text-embeddings-for-search-14-638.jpg?cb=1448540745"><img src="https://drive.google.com/uc?export=view&id=1vgCilYVR6LbIFNvu0A67srpOYEgzIcQm" width=75%></a>


**This is the dominant approach to building custom chatbots now.**

It utilizes **pre-trained embeddings** and minimal learning (only learning some distance based simple classification, even when super sophisticated, can be an "active learning" assisted simple model.) 

<a href="https://cdn-images-1.medium.com/max/1600/1*LVJuC7MeyWPj2ImkB3Xvwg.png"><img src="https://drive.google.com/uc?export=view&id=1g-o5PaCbREr4JXm9DvlnE6v66d99fbSm" width=55%></a>

The buildup of the "UX", that is the flow and all the responses are still tedious, but on the positive side, allow for detailed control.

This approach works well enough for a closed domain for production.

#### Problems

- It is tedious to build up
- The language production part is still "static", in agglutinative languages this in itslef can be problematic to some extent.

Naturally enough, the "end-to-end" approach is being actively developed here also.

# End-to-end methods


**Basic idea: nothing prevents us from formulating questions and answers as a seq2seq problem,** and learn it from a decently big corpus of real life conversations.

In case of big customer service operations, and _if_ logging is done properly (not always the case!) millions of dialogs are available, which from the quantity perspective is enough to train deep learning models. 

Before going into detail about problems effecting the end-to-end dialog systems, major caveat:
The ancient **["GIGO"](https://en.wikipedia.org/wiki/Garbage_in,_garbage_out)** principle applies, and we would not dare to think, how much of rubbish can be found in a customer service chatlog. (The curious case of "Juan", who always sends each customer away without even trying to answer questions...)

## Problem of (semantic) control in response generation

In case of end to end models, the goal is to generate interesting, varied and natural texts, but with a controlled content. In this case the "decoder" side is emphasized, and has to guarantee good quality output, and beside the "well formedness" of the sentences avoid the case when:


- The generated sentences do not have the desired content in them.
- They have it multiple times (like "stuttering"...)

### Specific architectures for semantic control of the output

[Semantically Conditioned LSTM-based Natural Language Generation for Spoken Dialogue Systems](https://arxiv.org/pdf/1508.01745.pdf)

<a href="https://ai2-s2-public.s3.amazonaws.com/figures/2017-08-08/0b3a0710031be11b2ef50437c7d9eb52c91d6a33/3-Figure1-1.png"><img src="https://drive.google.com/uc?export=view&id=1L8eF2NpaNmCtW-FH3SsSpvoFyKjGgvCy" width=45%></a>

## Heavy use of co-references

In real dialogs people tend to use much less explicit references than in written language, thus the number of unresolved co-references is much higher. (Remember Wittgenstein's language play? "Just give me that please!")

Since the resolution of co-references is the most obscure part of semantics, many times strong "common sense" background knowledge and/or "pragmatics" would be necessary to carry out this task.

As we well know, "common sense" is heavily under construction...


## (Long term) context


Interactions between humans and chatbots tend to be limited in length (which is a difference from human to human interactions.

<a href="https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcSrrpWBojmX96lyl1pbx5cM301oi0VQcpi93-XvJ2uyEodDQoHY"><img src="https://drive.google.com/uc?export=view&id=1TZPKKKL4S4wAH7L-W_-ASd0kOi9Q61e1" width=45%></a>

But even during these shorter conversations, humans tend to use a longer context window, pointing back to facts mentioned long ago, as well as sometimes switching contexts rapidly. These present fundamental challenges for dialog systems.

There are some forward looking solutions, see for example the very promising [Rasa.AI](https://rasa.com/) framework with ["context aware dialogs"](https://rasa.com/docs/nlu/0.12.3/context/).

## Need for monitoring in production

If these systems go live, typically they impact customers. At this point genuine care has to be taken to allow for:

- Monitoring
- Control
- Intervention (takeover by a human agent)

A production solution has to include logic for these requirements.
